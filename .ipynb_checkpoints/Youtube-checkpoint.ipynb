```python
from googleapiclient.discovery import build

# Ganti dengan API Key
API_KEY = ''
VIDEO_ID = 'UPrkC1LdlLY'

# Inisialisasi YouTube API client
youtube = build('youtube', 'v3', developerKey=API_KEY)

def get_comments(video_id, max_results=10):
    comments = []
    results = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        textFormat='plainText',
        maxResults=max_results
    ).execute()

    while results and len(comments) < max_results:
        for item in results.get('items', []):
            snippet = item['snippet']['topLevelComment']['snippet']
            comment = {
                'author': snippet['authorDisplayName'],
                'publishedAt': snippet['publishedAt'],
                'likeCount': snippet.get('likeCount', 'N/A'),
                'textOriginal': snippet['textOriginal']
            }
            comments.append(comment)

            # Hentikan jika sudah mendapatkan jumlah komentar yang diinginkan
            if len(comments) >= max_results:
                break

        # Periksa apakah ada komentar tambahan (pagination)
        if len(comments) < max_results and 'nextPageToken' in results:
            next_page_token = results['nextPageToken']
            results = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                pageToken=next_page_token,
                textFormat='plainText',
                maxResults=max_results - len(comments)  # Ambil sisanya dari yang dibutuhkan
            ).execute()
        else:
            break

    return comments

# Ambil komentar
comments = get_comments(VIDEO_ID, max_results=10)

# Simpan ke DataFrame dan kemudian ke CSV
df = pd.DataFrame(comments)
df.to_csv('youtube_comments.csv', index=False)

print("Komentar telah disimpan ke youtube_comments.csv")
```

```python
import pandas as pd

# Load data dari CSV
df = pd.read_csv('youtube_comments.csv')

# Tampilkan beberapa baris pertama untuk memastikan data telah dimuat
df.head()
```

```python
import re

# Fungsi untuk membersihkan teks
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Menghapus whitespace berlebihan
    text = re.sub(r'[^\w\s]', '', text)  # Menghapus tanda baca
    text = text.lower()  # Mengubah ke huruf kecil
    return text

# Terapkan pembersihan pada kolom komentar
df['clean_text'] = df['textOriginal'].apply(clean_text)

# Tampilkan beberapa baris setelah pembersihan
df[['textOriginal', 'clean_text']].head()
```

```python
from textblob import TextBlob

# Fungsi untuk mendapatkan sentimen
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Terapkan analisis sentimen pada kolom komentar
df['sentiment'] = df['clean_text'].apply(get_sentiment)

# Tampilkan hasil analisis sentimen
df[['textOriginal', 'sentiment']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Hitung jumlah komentar berdasarkan sentimen
sentiment_counts = df['sentiment'].value_counts()

# Visualisasi distribusi sentimen tanpa legend
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, hue=sentiment_counts.index, legend=False, palette='viridis')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Distribution of YouTube Comments Sentiment')
plt.show()
```

```python
# Simpan DataFrame dengan hasil analisis sentimen ke file CSV
df.to_csv('youtube_comments_with_sentiment.csv', index=False)

print("Hasil analisis sentimen telah disimpan ke youtube_comments_with_sentiment.csv")
```
